```python
test_pred = svd_model.test(testset)
```

```python
float(accuracy.rmse(test_pred))
```

```python
score_values = 100
for i in np.arange(2,20,2):
    svd_model = SVD(n_factors=i)
    svd_model.fit(trainset)
    test_pred = svd_model.test(testset)
    if score_values > float(accuracy.rmse(test_pred)):
        score_values = float(accuracy.rmse(test_pred))
        print(i, float(accuracy.rmse(test_pred)))
```

#### Creating a function which will recommend product based on the prediction provided by item based and used based collaborative filtering

```python
def top_5_recommendations(uid):
    all_products = df_data_1['productid'].unique()
    user_recom_df = pd.DataFrame()
    item_rating_icf = [] # Item rating by icf
    item_rating_ucf = [] # Item rating by bcf
    top_5_recom_by_icf = []
    top_5_recom_by_ucf =[]
    
    prod_list = [x for x in all_products if x not in df_data_1[df_data_1['authorid'] == uid]['productid'].unique()]
    
    for i in prod_list:
        item_rating_icf.append(icf_mdl.predict(uid, i).est)
        item_rating_ucf.append(ucf_mdl.predict(uid, i).est)
    
    user_recom_df = pd.DataFrame({'productid': prod_list, 'ICF_Est' : item_rating_icf, 'UCF_Est':item_rating_ucf})
    
    top_5_recom_by_icf = user_recom_df.sort_values(by='ICF_Est', ascending=False).head(5)['productid']
    top_5_recom_by_ucf = user_recom_df.sort_values(by='UCF_Est', ascending=False).head(5)['productid']
    
    print('Top 5 Recommendation with Item Based Collaborative Filter model are: ')
    for prd in top_5_recom_by_icf:
        print(df_data[df_data_1['productid'] == prd]['product'].unique())
        
    print('')

    print('Top 5 Recommendation with User Based Collaborative Filter model are: ')
    for prd in top_5_recom_by_ucf:
        print(df_data[df_data_1['productid'] == prd]['product'].unique())
    
```

## 10. Check for outliers and impute them as required.

Outliers can create bias in actual prediction. For example Fake profiles, users who gave high ratings to one product to influence recommendation. Outlier in Recommendation system are following records or users:
    1. One who gave almost same or similar rating for all products.
    2. User who tends to give less rating to all the products (their mean rating will be low as compared to all the users ) 
    3. User who tends to give very high  rating to all the products (their mean rating will be high as compared to all the users )
    4. Users who have given rating to very less products (one or 2)

```python
print('Total number of products : ',df_data_1['productid'].nunique())
print('Total number of author : ', df_data_1['authorid'].nunique())
```

```python
authorid_scr  = [] # This will cover our first three points (1,2,3)
authorid_prod = [] # This will cover last point (4)

for i in df_data_1['authorid'].unique():
    if df_data_1[df_data_1['authorid'] == i]['score'].nunique() < 3:
         authorid_scr.append(i) # Users who have provided only 1 or 2 type of rating
    if df_data_1[df_data_1['authorid'] == i]['productid'].nunique() < 6:
         authorid_prod.append(i) # Users who have provided rating for a few products (say 5 products only) out of 3336
            
print('Users who have provided one or two type of ratings only', len(authorid_scr))
print('Users who have provided ratings for less than 10 products', len(authorid_prod))
```

## 12. In what business scenario you should use popularity based Recommendation Systems?

Situation where popularity based model are good:
1. In situations where we do not have any data/information of user, we cannot create user similarity matrix.
2. An an anonymous user login to system
3. A user recently joined the the system
4. Grey sheep problems is present in system from a good time however user is very disimilar in terms of what he/she like/watch from all existing users.
5. It help us in cold start business problem.

## 13. In what business scenario you should use CF based Recommendation Systems?

Scenarios where Collborative Filter Based Recommendation system is useful:
1. It's a personalized Recommendation system based on similarity. We create user-user and item-item similarity matrix based on rating given by users. We will be able to use this system only after getting sufficient information about the users and items.
2. If we do not have information about the products contents, item-item similarity matrix can help to identify similarity between products.
3. More personalized Recommendation system, if a person like only sci-fi, action movies only with CF based will help recommending him such movies only. Similarly if a person like certain type of phone we will recommend only those type of phone.
4. Help us avoid situation of recommending product/item to a user completely different from what he/she like. This could be a situation with popularity based model.
5. Suffer in cold start problem.

## 14. What other possible methods can you think of which can further improve the recommendation for different users?

- We can improve recommendation by creating a hybrid model. Where we mix popularity based model with user based and item based Collaborative Filter recommendation models.
- Users who have recently joined and users who have given rating for a few products only, it's difficult to find similarity of user or items for such users, we can show popular product in such scenarios.
- Identify and remove outliers from the dataset, Remove fake user ratings -> Ratings which are same for all products (can be both low or high but same for all) add bias error in model. This need to be done for Collaborative filter recommendatio model.
- There could be certain products which have not rated by anyone yet or rated by a very few users. These products can be new products. For such products we can go for content based model and recommend similar content to user.
